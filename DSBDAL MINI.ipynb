### About the analysis

1. The above analysis was done on dataset consisting of students review about PICT website.
2. The website was judged upon various factors which are considered to be good design factors - UI, Color, Responsivness,    Accessibility.
3. The data was gathered using google form and contained more than responses more than 97 students. The form included 5 mcq question and one open ended question describing own review about PICT website.
4. As the data was collected through MCQ based formed, the dataset did not contain outliers.

### Conclusion about the Analysis

1. Making conclusion about the content the website provides , most of the students voted 4/5 . This states that the content of the website is satisfiable and the website provides good information and notices regarding college and events.
2. Students vote 2-3/5 for UI of the website, which signifies that students are not in the favour of old UI and demands change of it . Applying the sentiment analysis to open ended question below , we found out that students are not happy with the dark blue color of the website.
3. Students gave 3/5 for both accessibility & responsiveness of the website. This is the average rating for both the components of the website, and improvement is possible in these aspects. The accessibility and responsiveness of the website can be enhanced by placing the content in the right position according to heirarchy & using more media query tags respectively.
4. Color of the website got a very poor ratng from th student. This factor can be improved on a urgent basis looking at the need and taking into consideration the poor rating of students.
